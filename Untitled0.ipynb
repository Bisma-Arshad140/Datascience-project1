```python
df = df.loc[:,
            ~df.columns.str.contains('^zero')]
```

## step 1: Data Cleaning
we removed unecessary columns and renamed mislabled columns.

```python
df = df.rename(columns={'2urvived':'Survived'})
```

```python
df.columns
```

## Step2: Visualization Insights
- more females survived than males.
- 1st class paseengers had higher survival rates.

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived',data=df)
plt.title('Survival by Genedr')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived',data=df)
plt.title('Survival by Passenger Class')
plt.show()
```
